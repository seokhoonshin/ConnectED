```python
import pandas as pd
import matplotlib.pyplot as plt


column_name = "How often in the last 12 months did your mental health keep you from participating in your normal activities and relationships?"
belonging_column = "Belonging Barometer Number"  
gender_column = "What is your gender?" 

# Define gender categories
gender_list = ["female", "male"]

# Reorder the categories of days missed due to mental health
days_missed_order = ["it didn't", "a day or two", "a week or less", "a week or two", "two weeks to a month", "more than a month"]

# Convert the 'days missed' column to a Categorical type with the specified order
data[column_name] = pd.Categorical(data[column_name], categories=days_missed_order, ordered=True)

# Classify students based on their belonging barometer number with the custom ranges
data['Belonging Category'] = pd.cut(data[belonging_column], 
                                    bins=[0, 2.9, 5], 
                                    labels=['Low', 'High'], 
                                    right=True)

# Create an empty DataFrame to store the counts for both genders and belonging scores
combined_distribution = pd.DataFrame()

# Loop through each gender and belonging category, and store the counts
for gender in gender_list:
    for belonging_category in ['Low', 'High']:
        # Filter data for the specific gender and belonging category
        gender_belonging_students = data[(data[gender_column] == gender) & (data['Belonging Category'] == belonging_category)]
        
        # Count the frequency of each value in the 'days missed' column for the current gender and belonging category
        days_missed_distribution = gender_belonging_students[column_name].value_counts()
        
        # Add the distribution to the combined DataFrame with gender and belonging category as multi-index
        combined_distribution[(gender, belonging_category)] = days_missed_distribution

# Define the custom color palette
color_palette = ['#D7A100', '#4D6F38', '#4B3F29', '#A5B3B0']

# Plot the distribution of days missed due to mental health by gender and belonging category
combined_distribution.plot(kind='bar', figsize=(14, 8), color=color_palette[:len(combined_distribution.columns)])

# Title and labels
plt.title('Distribution of Days Missed Due to Mental Health by Gender and Belonging Score')
plt.xlabel('Days Missed Due to Mental Health')
plt.ylabel('Frequency')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a legend for better clarity
plt.legend(title='Gender and Belonging Score')

# Ensure labels fit
plt.tight_layout()

# Show the plot
plt.show()
```
